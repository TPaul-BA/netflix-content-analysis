# Importing Dependencies

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Loading Data

```python
# Loading Cleaned Data
df = pd.read_csv("../data/cleaned/netflix_cleaned.csv")

df.head()
```

```python
df.info()
```

# Analysis 1 - Movies vs TV Shows Trend

```python
sns.countplot(data = df, x = 'type')
plt.title("Content Type Distribution")
plt.show()
```

# Analysis 2 - Growth Over Time

```python
df.groupby('year_added')['show_id'].count().plot()
plt.title("Content Added Over Years")
plt.show()
```

# Analysis 3 - Top Genres

```python
df['primary_genre'].value_counts().head(10).plot(kind = 'bar')
plt.title("Top Genres")
plt.show()
```

# Analysis 4 - Country Analysis

```python
df['country'].value_counts().head(10)
```

# Analysis 5 - Duration Insights

```python
sns.histplot(df[df['type'] == 'Movie']['duration_int'], bins = 30)
plt.title("Movie Duration Distribution")
plt.show()
```
